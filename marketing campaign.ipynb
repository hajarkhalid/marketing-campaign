```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv(r"D:\marketing_data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.replace(' ', '')
df["Income"] = df["Income"].str.replace('$','')
df["Income"] = df["Income"].str.replace(',','').astype('float')
df
```

```python
df["Dt_Customer"]=pd.to_datetime(df["Dt_Customer"])
df
```

```python
df.isna().sum()
```

```python
ax = sns.boxplot(x="Income",
                 data= df, palette="tab20_r")
```

```python
plt.figure(figsize=(8,4))
sns.distplot(df["Income"],kde=False)
plt.title('income distribution')
plt.ylabel('count')
```

```python
df['Income'] = df['Income'].fillna(df['Income'].median()) 
df
```

```python
df1 = df.drop(columns=['ID', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'Complain']).select_dtypes(include=np.number)
df1
```

```python
df1.plot(subplots=True, layout=(4,4), kind='box', figsize=(12,14))
plt.subplots_adjust(wspace=0.5)
```

```python
df = df[df['Year_Birth'] > 1900].reset_index(drop=True)
plt.figure(figsize=(3,4))
df['Year_Birth'].plot(kind='box')
```

```python
df['Dependents'] = df['Kidhome'] + df['Teenhome']

# Year becoming a Customer
df['Year_Customer'] = pd.DatetimeIndex(df['Dt_Customer']).year

# Total Amount Spent
mnt_cols = [col for col in df.columns if 'Mnt' in col]
df['TotalMnt'] = df[mnt_cols].sum(axis=1)

# Total Purchases
purchases_cols = [col for col in df.columns if 'Purchases' in col]
df['TotalPurchases'] = df[purchases_cols].sum(axis=1)

# Total Campaigns Accepted
campaigns_cols = [col for col in df.columns if 'Cmp' in col] + ['Response'] # 'Response' is for the latest campaign
df['TotalCampaignsAcc'] = df[campaigns_cols].sum(axis=1)

# view new features, by customer ID
df[['ID', 'Dependents', 'Year_Customer', 'TotalMnt', 'TotalPurchases', 'TotalCampaignsAcc']].head()
```

```python
plt.figure(1 , figsize = (15 , 6))
n = 0 
for x in ['Year_Customer' , 'TotalPurchases' , 'TotalMnt','TotalCampaignsAcc','Dependents']:
    n += 1
    plt.subplot(1 ,5, n)
    plt.subplots_adjust(hspace =0.5 , wspace = 0.5)
    sns.distplot(df[x] , bins = 20)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
plt.figure(1 , figsize = (15 , 7))
n = 0 
for x in ['Year_Customer' , 'TotalPurchases' , 'TotalMnt','TotalCampaignsAcc','Dependents']:
    for y in ['Year_Customer' , 'TotalPurchases' , 'TotalMnt','TotalCampaignsAcc','Dependents']:
        n += 1
        plt.subplot(3 , 5 , n)
        plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
        sns.regplot(x = x , y = y , data = df)
        plt.ylabel(y.split()[0]+' '+y.split()[1] if len(y.split()) > 1 else y )
plt.show()
```

```python
plt.figure(1 , figsize = (15 , 7))
n = 0 
for cols in ['Year_Customer' , 'TotalPurchases' , 'TotalMnt','TotalCampaignsAcc','Dependents']:
    n += 1 
    plt.subplot(1 , 5 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.violinplot(x = cols , y = 'Dependents' , data = df , palette = 'vlag')
    sns.swarmplot(x = cols , y = 'Dependents' , data = df)
    plt.ylabel('Dependents' if n == 1 else '')
    plt.title('Boxplots & Swarmplots' if n == 2 else '')
plt.show()
```

```python
plt.figure(1 , figsize = (15 , 7))
n = 0 
for cols in ['Year_Customer' , 'TotalPurchases' , 'TotalMnt','TotalCampaignsAcc','Dependents']:
    n += 1 
    plt.subplot(1 , 5 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.violinplot(x = cols , y = 'Kidhome' , data = df , palette = 'vlag')
    sns.swarmplot(x = cols , y = 'Teenhome' , data = df)
    plt.ylabel('Dependents' if n == 1 else '')
    plt.title('Boxplots & Swarmplots' if n == 2 else '')
plt.show()
```
